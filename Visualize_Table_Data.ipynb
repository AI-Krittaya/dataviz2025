```python
import pandas as pd
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
display(iris_df.head())
```

```python
iris_df= iris_df.drop('target', axis=1)
```

# Heatmap

```python
from matplotlib import pyplot as plt
```

```python
import matplotlib
```

```python
matplotlib.rcParams['figure.figsize'] = [15, 10] #กำหนดค่าของ figure

plt.imshow(iris_df.iloc[:,:],interpolation='nearest')

#จากรูปแบ่งได้3กลุ่มดูจากสีที่เปลี่ยนไปแต่งละช่วง
```

# pandas: pararell coordinate

```python
iris = load_iris()
iris_df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
iris_df['target'] = iris.target
display(iris_df.head())
```

```python
matplotlib.rcParams['figure.figsize'] = [5, 5]
pd.plotting.parallel_coordinates(iris_df, 'target')

#ถ้า column เยอะจะดูกราฟได้ยากขึ้น ดีสำหรับ column น้อย
```

# pandas: scatter matrix

เพื่อดูการกระจายแต่จุดของข้อมูล column ต่างๆ

ตรงเส้นทแยงมุมเป็นกราฟแท่ง คือ ความสัมพันธ์ระหว่างตัวมันเอง จึงเป็นความถี่ที่แสดงออกมา

```python
pd.plotting.scatter_matrix(iris_df);
```

# Color

```python
matplotlib.rcParams['figure.figsize'] = [5, 10] #กำหนดค่าของ figure

plt.imshow(iris_df.iloc[:,:],interpolation='nearest', cmap='CMRmap')
```

```python
matplotlib.rcParams['figure.figsize'] = [5, 5]
pd.plotting.parallel_coordinates(iris_df, 'target', color=['red', 'green', 'blue']) #ตัวย่อชื่อสี c=ฟ้า m=ม่วง y=เหลือง k=ดำ r=แดง g=เขียว b=น้ำเงิน
```

```python
matplotlib.rcParams['figure.figsize'] = [5, 5]
pd.plotting.parallel_coordinates(iris_df, 'target', color=['m', 'b', 'c'])
```

```python
matplotlib.rcParams['figure.figsize'] = [5, 5]
pd.plotting.parallel_coordinates(iris_df, 'target', color=[[1,0,1],[0.5,0.5,0.5],[0,1,1]]) #ทำสี R G B
```

```python
matplotlib.rcParams['figure.figsize'] = [5, 5]
pd.plotting.parallel_coordinates(iris_df, 'target', color=['#D100FF','#0014FF','#A1FF00']) #ทำสี R G B
```
